## Análisis Tiempo-Frecuencia
### Tarea 2 

```julia
f1=50 #frec 1
f2=60 #frec 2
seg=10 #tiempo de muestreo
sps=140 #muestras por segundo
t=collect(linspace(0,seg,seg*sps)) #vector de tiempo
y=cos.(2*π*t*f1) + cos.(2*π*t*f2); 
```

```julia
using PyPlot
```

```julia
plot(y)
```

```julia
f=abs.(fft(y,1));
```

```julia
frecvec=collect(linspace(0,sps,sps*seg))
length(frecvec)
length(f)
```

```julia
subplot(2,1,1)
plot(frecvec[1:700],f[1:700],"-r")
xlabel("Frec [Hz]")
ylabel("Amplitud")
subplot(2,1,2)
plot(t,y)
```
